## Einführung in das Projekt

In diesem Projekt analysieren wir Daten der OECD zur Lebensqualität (Well-Being) in verschiedenen Ländern und über mehrere Jahre hinweg.  
Unser Ziel ist es zu untersuchen, wie sich Lebensqualität international unterscheidet und wie sie sich im Zeitverlauf entwickelt.

Der Datensatz enthält Indikatoren zur Lebensqualität in unterschiedlichen Bereichen, zum Beispiel:

- Gesundheit  
- Bildung  
- Einkommen  
- Arbeitsmarkt / Beschäftigung  
- Sicherheit  
- subjektive Lebenszufriedenheit  

Auf Basis dieser Daten möchten wir:

1. einen Überblick über die Lebensqualität in den OECD-Ländern gewinnen,  
2. wichtige Zusammenhänge zwischen einzelnen Indikatoren (z.B. Einkommen und Lebenszufriedenheit) untersuchen und  
3. beobachten, wie sich diese Größen über die Zeit verändern.

In den folgenden Notebooks wenden wir Schritt für Schritt die Inhalte aus der Vorlesung auf diesen Datensatz an.  
Jede Datei (01_Lecture bis 06_Lecture) knüpft dabei an eine Vorlesung an und vertieft einen bestimmten statistischen Aspekt (z.B. deskriptive Statistik, Zusammenhänge, Modelle).


## Erste Hypothesen

Bevor wir mit der eigentlichen Datenanalyse begonnen haben, haben wir uns überlegt, welche Zusammenhänge wir im Datensatz erwarten.

Wir gehen davon aus, dass sich in den Daten klare Muster erkennen lassen. Zum Beispiel:

- Länder mit höherem durchschnittlichem Einkommen haben **tendenziell eine höhere Lebenszufriedenheit**.  
- Länder mit höherer Arbeitslosigkeit weisen **tendenziell ein geringeres Wohlbefinden** auf.  

Im Verlauf des Projekts werden wir prüfen, inwieweit diese Hypothesen durch die Daten bestätigt oder widerlegt werden.
